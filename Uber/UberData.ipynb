# Data Preprocessing

```python
dataset['PURPOSE'].fillna("Not",inplace = True)
```

```python
dataset.head()
```

```python
dataset['START_DATE']=pd.to_datetime(dataset['START_DATE'],errors='coerce')
```

```python
dataset['END_DATE']=pd.to_datetime(dataset['END_DATE'],errors='coerce')
```

```python
dataset.info()
```

```python
from datetime import datetime
```

```python
dataset['Date']=pd.DatetimeIndex(dataset['START_DATE']).date
```

```python
dataset.head()
```

```python
dataset['Time']=pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

```python
dataset['Day-Night']=pd.cut(x=dataset['Time'],bins=[0,10,15,19,24],labels=['Morning','Afternoon','Evening','Night'])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.countplot(dataset['CATEGORY'])
```

```python
plt.subplot(1,2,2)
sns.countplot(dataset['PURPOSE'])
```

```python
sns.countplot(dataset['Day-Night'])
```

```python
dataset['Month']=pd.DatetimeIndex(dataset['START_DATE']).month
```

```python
dataset.head()
```

```python
dataset['Month']=pd.DatetimeIndex(dataset['START_DATE']).month
month_label={1.0: 'January' , 2.0:'February' , 3.0:'March' , 4.0:'April' ,5.0: 'May' ,6.0: 'June' , 7.0: 'July' , 8.0: 'August' , 9.0: 'September', 10.0: 'October' , 11.0:'November' , 12.0: 'December'}
dataset['Month'] = dataset.Month.map(month_label)
mon=dataset.Month.value_counts(sort=False)
```

```python
dataset.head()
```

```python
df=pd.DataFrame({
"Month":mon.values,
"Value count":dataset.groupby("Month",sort=False)['MILES'].max()
})
p= sns.lineplot(data=df)
p.set(xlabel="Month", ylabel="Value count")
```

```python
dataset['Week']=dataset.START_DATE.dt.weekday
week_label={0: 'Monday' , 1:'Tuesday' , 2:'Wednesday' , 3:'Thursday' ,4: 'Friday' ,5: 'Saturday' , 6: 'Sunday'}
dataset['Week'] = dataset['Week'].map(week_label)
```

```python
dataset.head()
```

```python
week_label=dataset.Week.value_counts()
week_order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
week_label = week_label.reindex(week_order, fill_value=0)
sns.barplot(x=week_label.index, y=week_label)
plt.xlabel('Week')
plt.ylabel('Count')
```
